Sử dụng Multiindex và Groupby

1.Tạo Multindex cho data frame cho stocks bằng cách sử dụng cột symbol và date làm chỉ mục

```python
import pandas as pd
import csv
```

```python
path = 'C:/Users/Admin/Documents/lesson1 Data science/Nguyen_Manh_Cuong_23174600007_DHKL17A1HN/DHKL17A1HN/Lab03/Data/stocks1.csv'
path1 = 'C:/Users/Admin/Documents/lesson1 Data science/Nguyen_Manh_Cuong_23174600007_DHKL17A1HN/DHKL17A1HN/Lab03/Data/stocks2.csv'
pd_stocks1 = pd.read_csv(path)
pd_stocks2 = pd.read_csv(path1)
```

```python
pd_stocks1.fillna(pd_stocks1['high'].mean(),inplace=True)
pd_stocks1
```

```python
stocks = pd.concat((pd_stocks1,pd_stocks2),ignore_index=True)
stocks
```

```python
stocks.set_index(['symbol', 'date'], inplace=True)
stocks
```

2. Sử dụng Groupby để đánh giá trung bình (open,high,low,close) và volume trung bình cho mỗi ngày và mỗi mã chứng khoán

```python
avg_value_symbol = stocks.groupby('symbol')[['open','high','low','close']].mean()
avg_value_symbol
```

```python
avg_value_date  = stocks.groupby('date')[['open','high','low','close']].mean()
avg_value_date
```
