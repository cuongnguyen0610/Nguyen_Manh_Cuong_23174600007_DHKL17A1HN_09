Lab3.2 xử lí dữ liệu null

1.Kiểm tra xem stocks1 có giá trị null nào không

```python
import csv
import pandas as pd
```

```python
path = 'C:/Users/Admin/Documents/lesson1 Data science/Nguyen_Manh_Cuong_23174600007_DHKL17A1HN/DHKL17A1HN/Lab03/Data/stocks1.csv'
stocks1 = []
with open(path,'r',encoding='utf-8') as file:
    read = csv.reader(file)

    header = next(read)

    data = list(read)

    for i,row in enumerate(data):
        for j,value in enumerate(row):
            if j==1:
                continue
            try:
                value_str = str(value).strip()
                if not value_str:
                    data[i][j] = None
                elif j==0:
                    continue
                
                elif '.' in value_str:
                    data[i][j] = float(value_str)
                else:
                    data[i][j] = int(value_str)
            except ValueError as e:
                print(f'Không chuyển đổi được giá trị "{value}" ở dòng {i+1}, cột {j+1}: {e}')

for i in data:
    stocks1.append(dict(zip(header,i)))
stocks1
```

```python
pd_stocks1 = pd.DataFrame(stocks1)
pd_stocks1
```

1.Kiểm tra xem có giá trị null nào không

```python
pd_stocks1.isnull().sum()
```

2.Thay thế giá trị null ở cột high bằng giá trị trung bình

```python
mean_high = pd_stocks1['high'].mean()
mean_high
```

```python
pd_stocks1.fillna(mean_high,inplace=True)
pd_stocks1
```

3.Thay giá trị null ở cột row bằng giá trị trung bình của row

```python
mean_low = pd_stocks1['low'].mean()
mean_low
```

```python
pd_stocks1.fillna(mean_low,inplace=True)
pd_stocks1
```
