1. Tạo Pavot table từ Data Frame stocks với date làm chỉ mục , symbol làm cột và giá trị trung bình của close làm giá trị


```python
import pandas as pd
import csv
```

```python
path = 'C:/Users/Admin/Documents/lesson1 Data science/Nguyen_Manh_Cuong_23174600007_DHKL17A1HN/DHKL17A1HN/Lab03/Data/stocks1.csv'
path1 = 'C:/Users/Admin/Documents/lesson1 Data science/Nguyen_Manh_Cuong_23174600007_DHKL17A1HN/DHKL17A1HN/Lab03/Data/stocks2.csv'
pd_stocks1 = pd.read_csv(path)
pd_stocks2 = pd.read_csv(path1)
```

```python
pd_stocks1.fillna(pd_stocks1['high'].mean(),inplace=True)
pd_stocks1
```

```python
stocks = pd.concat((pd_stocks1,pd_stocks2),ignore_index=True)
stocks
```

```python
pivot_place = stocks.pivot_table(values='close',index='date',columns='symbol',aggfunc='mean',)
pivot_place
```

2.Thêm 1 cột tính tổng volume giao dịch cho mỗi mã chứng khoán (symbol)

```python
stocks['total_volume'] = stocks.groupby('symbol')['volume'].transform(sum)
stocks
```

3.Sắp xếp pivot table trên tổng volume giao dịch ,từ cao xuống thấp

```python
pivot_table_1 = pd.pivot_table(stocks, values='volume',index='symbol', aggfunc='sum')
pivot_table_1
```

```python
sorted_value = pivot_table_1.sort_values(by='volume',ascending=False)
sorted_value
```
